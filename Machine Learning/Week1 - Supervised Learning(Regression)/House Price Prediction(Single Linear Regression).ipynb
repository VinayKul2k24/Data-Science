# <h1 align = center> **Training Model** </h1>

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("home_prices.csv")
df.head()
```

```python
model = LinearRegression()
```

```python
model.fit(df[["area_sqr_ft"]] , df["price_lakhs"])
```

# <h1 align = center> **Model Prediction** </h1>

```python
model.predict([[1500],[400]])
```

```python
model.coef_, model.intercept_
```

```python
price = model.coef_[0]*1500 + model.intercept_
```
